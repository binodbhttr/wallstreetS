# DBSCAN and HDBSCAN with examples

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
```

Density-Based Spatial Clustering of Applications with Noise (DBSCAN) is a clustering algorithm that groups together points that are close to each other in a high-dimensional space. It works by identifying clusters of points that are denser than their surroundings, and can handle clusters of arbitrary shape and size.

The algorithm requires two parameters: epsilon (Îµ) and minimum points (minPts). Epsilon is a distance parameter that defines the maximum distance between two points for them to be considered part of the same cluster, while minPts is the minimum number of points required for a group of points to be considered a cluster.

Here's an example of how DBSCAN clustering works. Suppose we have a dataset of 2-dimensional points:


[(1, 2), (2, 2), (3, 3), (3, 4), (4, 4), (6, 6), (7, 7), (8, 8)]


```python
a=[(1, 2), (2, 2), (3, 3), (3, 4), (4, 4), (6, 6), (7, 7), (8, 8)]
b=np.array(a)
print(a)

plt.scatter(b[:, 0], b[:, 1])
```

We can apply DBSCAN clustering to this dataset with epsilon = 1.5 and minPts = 3. Starting with a random point, the algorithm finds all points within a distance of epsilon, and recursively expands the cluster to include all points that are within epsilon of any point in the cluster, until no more points can be added. If the resulting cluster has fewer than minPts points, it is classified as noise.

In this example, the algorithm would start with the point (1, 2) and identify the points (2, 2) and (3, 3) as being within epsilon. It would then expand the cluster to include (3, 4) and (4, 4), but would stop there, since (6, 6) is farther than epsilon from the cluster. This results in a cluster of size 4:

```python
#Testing the method

dbscan = DBSCAN(eps=1.5, min_samples=3)
labels = dbscan.fit_predict(b)
plt.scatter(b[:, 0], b[:, 1], c=labels,cmap='viridis')
plt.show()
```

The algorithm would then start with the point (6, 6) and identify the points (7, 7) and (8, 8) as being within epsilon. This would result in a second cluster of size 3:

# Another example

```python
# Generate a 2-dimensional dataset of 100 points with 4 clusters
X, y = make_blobs(n_samples=1000, centers=3, n_features=2, random_state=42)

# Apply DBSCAN clustering with epsilon = 1 and minPts = 5
dbscan = DBSCAN(eps=1, min_samples=5)
labels = dbscan.fit_predict(X)
```

The fit_predict method returns an array of cluster labels for each point in the dataset.

Finally, we plot the resulting clusters using matplotlib. 
The scatter function takes the x and y coordinates of each point in the dataset, and the cluster labels as the color of each point. This results in a scatter plot where each cluster is represented by a different color.

```python
# Plot the resulting clusters
plt.scatter(X[:, 0], X[:, 1], c=labels,cmap='viridis')
plt.show()
```

# HDBSCAN

```python
import hdbscan
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Generate a random dataset
X, y = make_blobs(n_samples=1000, centers=3, random_state=42) #X is the data. y is the labels we already know

# Initialize the HDBSCAN algorithm with the desired parameters
hdbscan_alg = hdbscan.HDBSCAN(min_cluster_size=10, min_samples=5)

# Fit the algorithm to the data
hdbscan_alg.fit(X)

# Extract the cluster labels and probabilities for each point
labels = hdbscan_alg.labels_
probabilities = hdbscan_alg.probabilities_

# Plot the results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title('HDBSCAN Clustering')
plt.show()
```

In this example, we first generate a random dataset of 1000 points with 3 clusters using the make_blobs() function from scikit-learn.

We then initialize the HDBSCAN algorithm with two parameters: min_cluster_size and min_samples. The min_cluster_size parameter specifies the minimum number of points required for a cluster to be considered valid, while the min_samples parameter specifies the minimum number of points required for a point to be considered a core point.

We fit the algorithm to the data using the fit() method, and then extract the cluster labels and probabilities for each point using the labels_ and probabilities_ attributes.

Finally, we plot the results using a scatter plot, where the color of each point represents its assigned cluster label.

# How DBSCAN differs from fof ?

DBSCAN and friends of friends (FOF) are both density-based clustering algorithms, but they have some key differences:

Distance metric: DBSCAN uses a distance metric (e.g., Euclidean distance) to measure the distance between points, while FOF uses a graph-based approach to measure the connectivity between points. Specifically, FOF constructs a graph where each point is connected to its "friends" (i.e., other points that are within a certain distance threshold), and then clusters the graph using graph-based clustering algorithms such as the Louvain method.

Cluster shape: DBSCAN is capable of identifying clusters of arbitrary shape, while FOF is better suited for identifying clusters that are approximately spherical. This is because FOF uses a fixed distance threshold to determine which points are connected, which can result in clusters that are elongated or irregular in shape being split into multiple clusters.

Parameter tuning: DBSCAN requires the user to specify two parameters: eps, which controls the distance threshold for determining point connectivity, and min_samples, which controls the minimum number of points required to form a cluster. FOF also requires the user to specify a distance threshold, but the optimal threshold can often be determined using heuristics or trial-and-error.

Noise handling: DBSCAN is able to identify noise points (i.e., points that do not belong to any cluster) by labeling them with a special cluster label. FOF does not have a built-in mechanism for handling noise points, but they can be identified by looking for points that are not part of any cluster.

In summary, DBSCAN and FOF are both density-based clustering algorithms, but they differ in their approach to measuring distance/connectivity, their ability to handle different cluster shapes, their parameter tuning requirements, and their handling of noise points.

# Another HDBSCAN example

```python
import numpy as np
import matplotlib.pyplot as plt
import hdbscan
from sklearn.datasets import make_moons

# Generate a sample dataset with two moons
X, y = make_moons(n_samples=200, noise=0.05, random_state=0)

# Apply HDBSCAN clustering
clusterer = hdbscan.HDBSCAN(min_cluster_size=5,min_samples=5,cluster_selection_epsilon=0.005)
cluster_labels = clusterer.fit_predict(X)

# Plot the clustering result
plt.scatter(X[:,0], X[:,1], c=cluster_labels, cmap='rainbow')
plt.show()
```
